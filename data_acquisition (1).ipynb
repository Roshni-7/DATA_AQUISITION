```python
#Q9: Based on your analysis, provide a brief summary of any insights or patterns you discovered in the dataset.
import matplotlib.pyplot as plt

data.plot.scatter('Duration (hours)', 'Rating (out of 5)', color='gray')
plt.title('Course Duration vs. Price')
plt.xlabel('Duration (hours)')
plt.ylabel('Price ($)')
plt.figure(figsize=(10, 6))
plt.show()
```

```python
data.plot.bar('Category', 'Price ($)', color='blue')
plt.title('Category vs. Price')
plt.xlabel('Category')
plt.ylabel('Price ($)')
plt.figure(figsize=(10, 6))
plt.show()
```

```python
data.plot.box(subplots=True, figsize=(15, 10),layout=(3, 3))
plt.suptitle('Box Plots online insights')
plt.show()
```
